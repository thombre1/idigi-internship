```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from pylab import *

#deviation from mean
#x is a list of data
def de_mean(x):
    xmean = mean(x)
    return [xi-xmean for xi in x]

# How much different these two data set are: x be pageSpeed and y be purchaseAmount
def covariance(x,y):
    n = len(x)
    # sample covariance as we are dividing by n-1
    # dot is dotproduct of all the values of x and y
    return dot(de_mean(x), de_mean(y)) / (n-1)

# target mean, spread(Standard Deviation), No of Samples
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
# purchaseAmount = np.random.normal(50.0, 10.0, 1000)/pageSpeeds
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

scatter(pageSpeeds, purchaseAmount)
print(covariance(pageSpeeds, purchaseAmount))

plt.show()
```

```python
# Numpy has its own correlation function so that we dont have
# to do the above mathematics ourselves
# where we first calculate the mean and store the difference form
# mean in a list and then pass the lists to the cpvariance funtion
# which does the calculations by calculating the dot product and 
# dividing them by n-1 because its sample covariance
np.corrcoef(pageSpeeds, purchaseAmount)
```
