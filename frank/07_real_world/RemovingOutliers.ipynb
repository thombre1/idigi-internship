```python
%matplotlib inline
import numpy as np

incomes = np.random.normal(27000, 15000, 10000)
incomes = np.append(incomes, [np.random.normal(40000000,150000, 10)])
```

```python
import matplotlib.pyplot as plt

plt.hist(incomes, 100)
plt.show()
```

```python
def remove_outliers(data):
    u = np.mean(data)
    s = np.std(data)
    filtered_data = [e for e in data if (u - 2*s < e < u + 2*s)]
    return filtered_data
```

```python
filtered_data = remove_outliers(incomes)
plt.hist(filtered_data, 100)
plt.show()
```
