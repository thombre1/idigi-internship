### Similarities of Movies

```python
import pandas as pd 

r_cols = ['user_id', 'movie_id', 'rating']
#So, u.data seems to be a data format
# sep means a delimiter '\t'
# names = r_cols, means use the list provided for names of column
# usecols = 0,1,2
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

# Seemes like a Join Function - Joins with the movie_id
ratings = pd.merge(movies, ratings)
ratings.head(5)
```

```python
movieRatings = ratings.pivot_table(index=['user_id'], columns=['title'], values='rating')
movieRatings.head()
```

```python
starWarsRating = movieRatings['Star Wars (1977)']
starWarsRating.head()
```

```python
similarMovies = movieRatings.corrwith(starWarsRating).dropna()
df = pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False)
```

```python
import numpy as np
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

```python
#Movies that more than 100 people have rated
popularMovies = movieStats['rating']['size'] >= 100
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False).head(15)
```

```python
mappedColumnMoviestat = movieStats[popularMovies]

#some list comprehension, every element is a formatted string with i and j
# put the element if j is not empty(size is non zero that means someone has rated the movie)
# if j is empty then only keep i(rating) without the pipe |

mappedColumnMoviestat.columns=[f'{i}|{j}' if j!= '' else f'{i}' for i,j in mappedColumnMoviestat.columns]
df = mappedColumnMoviestat.join(pd.DataFrame(similarMovies, columns=['similarity']))
df.head()
```

```python
df.sort_values(['similarity'], ascending=False)
```
